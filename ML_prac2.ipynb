```python
data = pd.read_csv('C://Users/Admin/Downloads/DecisionTreeData.csv')
```

```python
X = data.iloc[:,:-1]
```

```python
X
```

```python
y = data.iloc[:,-1]
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
X['Age'] = le.fit_transform(X['Age'])
X['Income'] = le.fit_transform(X['Income'])
X['Gender'] = le.fit_transform(X['Gender'])
X['Marital Status'] = le.fit_transform(X['Marital Status'])
X
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(criterion='entropy', max_depth=None, min_samples_leaf=4, random_state=0)
```

```python
model = clf.fit(X.iloc[:,1:5],y)
```

```python
model
```

```python
X.iloc[:,1:5]
```

```python
y_pred = [[1,1,0,0]]

#model.predict(y_pred)

print("Prediction is: ", model.predict(y_pred))
```

```python
model.score(X.iloc[:,1:5],y) * 100
```
